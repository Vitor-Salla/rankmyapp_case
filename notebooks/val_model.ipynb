### M√©tricas

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error


df = df.dropna() # est√° com uma linha nula

# overall metrics
y_true = df['dauReal']
y_pred = df['forecast']

mae  = mean_absolute_error(y_true, y_pred)
mse  = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
mape = (np.abs((y_true - y_pred) / y_true).mean()) * 100    # in percent

print(f"MAE: {mae:.0f}")
print(f"RMSE: {rmse:.0f}")
print(f"MAPE: {mape:.1f}%")
```

```python
# comparando com um naive forecast (m√©dia por app)

y_true = df['dauReal']
y_pred = df['dau_avg']

mae  = mean_absolute_error(y_true, y_pred)
mse  = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
mape = (np.abs((y_true - y_pred) / y_true).mean()) * 100    # in percent

print(f"MAE: {mae:.0f}")
print(f"RMSE: {rmse:.0f}")
print(f"MAPE: {mape:.1f}%")
```
